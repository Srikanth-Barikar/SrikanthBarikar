```python
df.head()
import warnings
warnings.filterwarnings('ignore')
del df['CarName']
```

```python
df2 = pd.get_dummies(df, columns=['carCompany','enginelocation','fueltype','aspiration','doornumber','carbody','drivewheel','enginetype','cylindernumber','fuelsystem','symboling'])
df2
```

```python
#Training the MODEL
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df_train, df_test = train_test_split(df2, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
scaler = MinMaxScaler()
```

```python
cols = ['wheelbase', 'carwidth', 'curbweight','enginesize','boreratio','horsepower','price']
df_train[cols] = scaler.fit_transform(df_train[cols])
```

```python
df_train.head()
#del df['CarName']
```

```python
#Dividing into X and Y sets for building the model
```

```python
ytrain = df_train.pop('price')
xtrain = df_train
```

```python
#Building Models using Linear Regression functions
```

```python
# LinearRegression and Importing RFE 
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
lrm = LinearRegression()
lrm.fit(xtrain, ytrain)
rfe = RFE(lrm, 12)
rfe = rfe.fit(xtrain, ytrain)
```

```python
list(zip(xtrain.columns,rfe.support_,rfe.ranking_))
```

```python
col1 = xtrain.columns[rfe.support_]
col1
```

```python
xtrain.columns[~rfe.support_]
```

```python
#Models using STAT MODELS
```

```python
#MODEL1
```

```python
xtrain_rfe = xtrain[col1]
```

```python
import statsmodels.api as sm 
xtrain_rfe = sm.add_constant(xtrain_rfe)
```

```python
lrm = sm.OLS(ytrain,xtrain_rfe).fit()
```

```python
print(lrm.summary())
```

```python
#-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*
```

```python
#Model 2
```

```python
xtrain1 = xtrain_rfe.drop(["enginelocation_front"], axis = 1)
xtrain_lrm = sm.add_constant(xtrain1)
```

```python
lrm = sm.OLS(ytrain,xtrain_lrm).fit()
```

```python
print(lrm.summary())
```

```python
#-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*-------*
```

```python
#MODEL 3
```

```python
xtrain2 = xtrain_rfe.drop(["curbweight"], axis = 1)
xtrain_lrm = sm.add_constant(xtrain2)
```

```python
lrm = sm.OLS(ytrain,xtrain_lrm).fit()
```

```python
print(lrm.summary())
```

```python
ytrain_price = lrm.predict(xtrain_lrm)
```

```python
fig = plt.figure()
sns.distplot((ytrain - ytrain_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)
```

```python
xtrain_lrm.columns
```

```python
xtrain_lrm = xtrain_lrm.drop(['const'], axis=1)
```
